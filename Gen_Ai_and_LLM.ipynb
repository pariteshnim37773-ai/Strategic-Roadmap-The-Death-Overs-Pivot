```python
import google.generativeai as genai

genai.configure(api_key="")
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")

response = model.generate_content("What is Artificial Intelligence?")

print(response.text)
```

```python
context = """
Sports are organized, competitive physical activities
or games that improve physical fitness, skills, and mental well-being,
providing entertainment for spectators. They range from individual activities
like running to team sports like football and cricket,
played both indoors and outdoors. Benefits include improved cardiovascular
health, reduced disease risk, enhanced social skills,
and increased mental toughness.
"""

question = "Explain Sports"

response = model.generate_content(f"Context: {context} Question: {question}")

print(response.text)
```

```python
domain = "education"
question = "Explain Neural networks"

response = model.generate_content(
    f"You are expert in{domain}. Answer: {question}"
)
print(response.text)
```
